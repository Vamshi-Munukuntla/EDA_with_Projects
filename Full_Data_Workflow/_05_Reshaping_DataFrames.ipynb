## Pivot Table

```python
table2.head()
```

```python
table2.pivot_table(values='Count',
                   index='Country',
                   columns='Medal',
                  aggfunc='sum',fill_value=0)
```

```python
table2.groupby(['Country','Medal']).Count.sum().unstack(fill_value=0)
```

```python
table2.pivot_table(values='Count',
                    index='Country',
                    columns='Medal',
                    aggfunc='sum',
                    fill_value=0,
                    margins=True,
                    margins_name='Total')
```

```python
agg = table2.groupby(['Country','Medal']).Count.sum().unstack(fill_value=0)
```

```python
agg['All'] = agg.sum(axis=1)
```

```python
agg
```

```python
agg.loc['All']  = agg.sum()
```

```python
agg
```

## pd.crosstab()

```python
titanic.head()
```

```python
titanic.info()
```

```python
pd.crosstab(titanic.sex,titanic.pclass,
            margins=True,
           normalize=True)*100
```

```python
titanic.groupby(['sex','pclass']).pclass.count().unstack()
```

```python
pd.crosstab(index=titanic.sex,
            columns=titanic.pclass,
            values=titanic.age,
           aggfunc='mean')
```

```python
titanic.head(1)
```

```python
titanic.pivot_table(index=['sex'],
                    columns='pclass',
                    values='age',
                   aggfunc='mean')
```

```python
pd.crosstab(index=titanic.sex,
           columns=titanic.pclass,
           values=titanic.fare,aggfunc='sum',
           margins=True)
```

```python
pd.crosstab(index=titanic.sex,
           columns=titanic.pclass,
           values=titanic.fare,aggfunc='sum',
           margins=True,normalize=True)*100
```

```python
pd.crosstab(index=titanic.sex,
           columns=titanic.pclass,
           values=titanic.fare,aggfunc='count')
```

```python
summer =pd.read_csv('datasets/summer.csv')
```

```python
summer.head(3)
```

```python
summer.info()
```

```python
pd.crosstab(index=[summer.Year,summer.Country],
           columns=summer.Medal,aggfunc='count',
           values=summer.Athlete,margins=True).fillna(0)
```

```python
summer.pivot_table(index=['Year','Country'],
                   values='Athlete',
                   aggfunc='count',columns='Medal',
                  margins=True,fill_value=0)
```

## Melting DataFrames with melt()

```python
table = pd.read_csv("datasets/table_2012.csv")
```

```python
table.head()
```

```python
table.shape
```

```python
melt_table = table.melt(id_vars='Country',
           value_vars=['Gold','Silver','Bronze'],
          var_name='Medal',value_name='Count')
```

```python
melt_table
```

```python
melt_table.pivot(index='Country',
                 columns='Medal',
                 values='Count').sort_values("Gold",ascending=False)
```
