# Advanced Statistical Plotting with Seaborn

## First Steps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
titanic = pd.read_csv("datasets/titanic.csv")
```

```python
titanic.head()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2,palette='viridis')
sns.countplot(data=titanic,x='sex',hue='pclass',)
plt.show()
```

## Categorical plots

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2,palette='viridis')
sns.countplot(data=titanic,x='sex',hue='pclass',)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.stripplot(data=titanic,x='sex',y='age',jitter=True,dodge=True,hue='pclass')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.stripplot(data=titanic,x='sex',y='age',
              jitter=True,dodge=True,hue='survived')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.swarmplot(data=titanic,x='sex',y='age',dodge=True,hue='pclass')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.violinplot(data=titanic,x='sex',y='age',dodge=True,hue='pclass')
sns.swarmplot(data=titanic,x='sex',y='age',
              dodge=True,hue='pclass',color='black',size=4.5)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.violinplot(data=titanic,x='pclass',y='age',dodge=True,hue='sex')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.violinplot(data=titanic,x='pclass',y='age',
               dodge=True,hue='sex',split=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.barplot(data=titanic,x='pclass',y='age',dodge=True,hue='sex',ci=95)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.pointplot(data=titanic,x='pclass',y='age',dodge=True,hue='sex',ci=95)
plt.show()
```

## Jointplots/ Regression

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.jointplot(data=titanic,x='age',y='fare',kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.jointplot(data=titanic,x='age',y='fare',kind='hex')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.jointplot(data=titanic,x='age',y='fare',kind='kde')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.jointplot(data=titanic,x='age',y='fare',kind='reg')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.lmplot(data=titanic,x='age',y='fare', aspect=1,height=8,row='sex')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.lmplot(data=titanic,x='age',y='fare', aspect=1,height=8,hue='pclass')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.lmplot(data=titanic,x='age',y='survived', aspect=1,logistic=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.lmplot(data=titanic,x='age',y='survived',col='pclass',
           hue='sex',aspect=1,logistic=True)
plt.show()
```

## Matrix plots / Heatmaps

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass),
           annot=True,fmt='d',cmap='Reds',vmax=150)
plt.show()
```

```python
pd.crosstab(titanic.sex, titanic.survived)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.heatmap(pd.crosstab(titanic.sex, titanic.survived),
           annot=True,fmt='d',cmap='Reds',vmax=None)
plt.show()
```

```python
pd.crosstab(titanic.sex, titanic.pclass,
            values=titanic.survived, aggfunc='mean')
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass,
                        values=titanic.survived, aggfunc='mean'),
           annot=True,cmap='Reds',vmax=None)
plt.show()
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.25)
sns.heatmap(titanic.corr(),annot=True,cmap='Reds',vmax=None)
plt.show()
```
