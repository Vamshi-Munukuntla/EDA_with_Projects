```python
def convert_object_to_category(df):
    """
    Converts the column with dtype 'object' to 'category'
    
    1. It's memory efficient
    2. columns with category dtype are more computationally 
       faster than object type during groupby operations
    """
    for column in df.columns:
        if df[column].dtypes == 'O':
            df[column] = df[column].astype('category')
    return df.dtypes
```
