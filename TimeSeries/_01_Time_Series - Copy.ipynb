## Downsampling Time Series with resample()

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
temp = pd.read_csv('temp.csv',parse_dates=['datetime'],index_col='datetime')
```

```python
temp.head()
```

```python
temp.info()
```

```python
list(temp.resample('D'))[0][1]
```

```python
temp.resample("D").first()
```

```python
temp.resample("D").last()
```

```python
temp.resample("D").mean()
```

```python
temp.resample("D").median()
```

```python
temp.resample("D").sum()
```

```python
temp.resample("2H").first()
```

```python
temp.resample('W').mean()
```

```python
temp.resample('W-Wed').mean()
```

```python
temp.resample('M').mean()
```

```python
temp.resample('MS').mean()
```

```python
temp.resample('MS',loffset='14D').mean()
```

```python
temp.resample("Q").mean()
```

```python
temp.resample("Q-Mar").mean()
```

```python
temp.resample("Y").mean()
```

```python
temp.resample("YS").mean()
```

```python
temp.resample("M", kind='period').mean()
```

```python
temp.resample('W',kind='period').mean()
```

```python
temp.resample("Q",kind='period').mean()
```

```python
temp.resample('A',kind='period').mean()
```

```python
temp_monthly = temp.resample("M",kind='period').mean()
```

```python
temp_monthly
```

```python
temp_monthly.info()
```

```python
temp_monthly.index
```

```python
temp_monthly.plot(figsize=(15,8), fontsize=15)
plt.show()
```

```python
temp.plot(figsize=(15,8), fontsize=15)
plt.show()
```
